```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([12,23,34,45,56,67,78,89,95])
y = np.array([11,25,31,48,52,63,79,84,91])
z = np.array([90,85,75,65,54,43,32,21,14])
```

## Line Plot

```python
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'o', mfc = 'r', ls = '--', color = 'black');
# markers => >, <, ^, D, o
```

```python
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, z, marker = 'o', mfc = 'r', ls = '--', color = 'black');
```

```python
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y,label = '2023', marker = 'o', mfc = 'r', ls = '--', color = 'black');
plt.plot(x, z,label = '2024', marker = 'o', mfc = 'blue', ls = '--', color = 'black');
plt.grid()
plt.legend();
```

```python
# subplot

plt.subplot(2,2,1)
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'o', mfc = 'r', ls = '--',
         color = 'black');

plt.subplot(2,2,2)
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, z, marker = 'o', mfc = 'r', ls = '--',
         color = 'black');
```

## Bar Plot

```python
courses = ['Python', 'Java', 'R', 'Ruby', 'Kotlin']
students = [75, 69, 45, 56, 60]
```

```python
plt.title('Student Enrollment')
plt.xlabel('Courses')
plt.ylabel('No of students')
plt.bar(courses, students, color = 'black', width = 0.3, 
       edgecolor = 'red');
plt.grid()
```

## Scatter Plot

```python
plt.title('Scatter Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.scatter(x, y, marker = '>');
```

## Pie Chart

```python
plt.pie(students, labels = courses, 
        autopct = '%.2f%%');
```

## Boxplot

```python
marks = [67,65,57,61,66,58,56,60,70,85]
```

```python
plt.boxplot(marks);
```

```python
# plot the subplot for
# line plot(combined), bar plot, scatter plot, pie chart
plt.figure(figsize=(16,10))

plt.subplot(2,2,1)
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y,label = '2023', marker = 'o', mfc = 'r', ls = '--', color = 'black');
plt.plot(x, z,label = '2024', marker = 'o', mfc = 'blue', ls = '--', color = 'black');
plt.grid()
plt.legend();

plt.subplot(2,2,2)
plt.title('Student Enrollment')
plt.xlabel('Courses')
plt.ylabel('No of students')
plt.bar(courses, students, color = 'black', width = 0.3, 
       edgecolor = 'red');
plt.grid()

plt.subplot(2,2,3)
plt.title('Scatter Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.scatter(x, y, marker = '>');

plt.subplot(2,2,4)
plt.pie(students, labels = courses, 
        autopct = '%.2f%%');

plt.savefig('My first plot')
```
