```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([12,23,34,45,55,64,78,86,90])
y = np.array([23,28,34,45,53,69,72,88,93])
z = np.array([98,87,76,54,45,34,21,16,9])
```

## Line Plot

```python
plt.title('Line Plot 1')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x , y, marker = 'o', mfc = 'black', ls = '--');
# markers => >, <, ^, o, D
```

```python
plt.title('Line Plot 2')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x , z, marker = 'o', mfc = 'r', ls = '--', color = 'green');
# markers => >, <, ^, o, D
```

```python
plt.title('Line Plot 1')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x , y, marker = 'o', mfc = 'black', ls = '--')
plt.plot(x , z, marker = 'o', mfc = 'r', ls = '--', color = 'green');
```

## Subplot

```python
plt.subplot(2,2,1)
plt.title('Line Plot 1')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x , y, marker = 'o', mfc = 'black', ls = '--');

plt.subplot(2,2,2)
plt.title('Line Plot 2')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x , z, marker = 'o', mfc = 'r', ls = '--', color = 'green');
```

## Bar Plot

```python
courses = ['Java', 'Python', 'Kotlin', 'R', 'Ruby']
students = [70, 72, 65, 45, 54]
```

```python
plt.title('Student Enrollment')
plt.xlabel('Courses')
plt.ylabel('No of Students')
plt.bar(courses, students, color = 'black', width = 0.4, edgecolor = 'red');
```

## Scatter Plot

```python
plt.scatter(x, y);
```

## Box plot

```python
marks = [56,67,61,60,98,45]
```

```python
plt.boxplot(marks);
```

## Pie Chart

```python
plt.pie(students, labels = courses, autopct = '%.2f%%');
```

```python
# combined line plot, pie chart, bar plot, scatter plot
# plot subplot of all of this
```

```python
plt.figure(figsize=(16,11))

plt.subplot(2,2,1)
plt.title('Line Plot 1')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x , y, marker = 'o', mfc = 'black', ls = '--')
plt.plot(x , z, marker = 'o', mfc = 'r', ls = '--', color = 'green');

plt.subplot(2,2,2)
plt.title('Student Enrollment')
plt.xlabel('Courses')
plt.ylabel('No of Students')
plt.bar(courses, students, color = 'black', width = 0.4, edgecolor = 'red');

plt.subplot(2,2,3)
plt.scatter(x, y);

plt.subplot(2,2,4)
plt.pie(students, labels = courses, autopct = '%.2f%%');

plt.savefig('My Fist Plot')
```
