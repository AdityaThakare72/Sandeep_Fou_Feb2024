## Dataframe Concate

```python
# create a dataframe with roll, gender and cgpa column
# in roll have roll no => 101 to 110
```

```python
cgpa = df['marks'] / 10 + 1.5
gender = ['m', 'm', 'm', 'm', 'f', 'f', 'm', 'f', 'm', 'm']
```

```python
newdf = pd.DataFrame({
    'roll': range(101,111),
    'gender': gender,
    'cgpa': cgpa
})
```

```python
newdf
```

```python
pd.merge(df, newdf, on = 'roll')
```
